```python
import PyPDF2

def get_pdf_page_count(pdf_file_path):
  """
  PDF 파일의 페이지 수를 반환하는 함수입니다.

  Args:
    pdf_file_path: PDF 파일의 경로

  Returns:
    PDF 파일의 페이지 수
  """

  with open(pdf_file_path, 'rb') as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    page_count = len(pdf_reader.pages)
    return page_count

# PDF 파일 경로
pdf_file_path = './대한민국헌법(헌법)(제00010호)(19880225).pdf'

# 페이지 수 가져오기
page_count = get_pdf_page_count(pdf_file_path)

# 결과 출력
print(f"PDF 파일 '{pdf_file_path}'의 페이지 수: {page_count}")
```

### 오류 처리: PDF 파일을 읽거나 파싱하는 과정에서 발생할 수 있는 오류를 처리하는 코드를 추가할 수 있습니다.

- 정규 표현식은 복잡한 패턴을 찾는 데 유용하지만, 잘못 사용하면 예상치 못한 결과를 초래할 수 있습니다.

```python
import PyPDF2
import re

def convert_pdf_to_markdown(pdf_text):
    """
    PDF 파일의 텍스트를 마크다운 형식으로 변환하는 함수입니다.

    Args:
        pdf_text: PDF 파일에서 추출한 텍스트 문자열

    Returns:
        마크다운 형식으로 변환된 텍스트 문자열
    """

    # 1. 줄 바꿈 처리: 문단 구분을 위해 줄 바꿈을 두 줄로 변환
    markdown_text = pdf_text.replace('\n', '\n\n')

    # 2. 제목 처리: '#' 기호를 사용하여 제목 레벨을 표시
    markdown_text = re.sub(r'^(#{1,6})\s+(.+)', r'\1 \2', markdown_text)

    # 3. 목록 처리: '-' 기호를 사용하여 목록 항목을 표시
    markdown_text = re.sub(r'^\s*-\s+(.+)', r'- \1', markdown_text)

    # 4. 표 처리: '|' 기호를 사용하여 표를 표시 (단순한 표만 처리)
    markdown_text = re.sub(r'\|(.+)\|', r'| \1 |', markdown_text)

    # 5. 강조 처리: '**' 기호를 사용하여 텍스트를 강조 표시
    markdown_text = re.sub(r'\*(.+)\*', r'**\1**', markdown_text)

    return markdown_text

def parse_pdf_to_text(pdf_file_path):
    """
    PDF 파일을 파싱하여 텍스트 데이터를 추출하는 함수입니다.

    Args:
        pdf_file_path: PDF 파일의 경로

    Returns:
        PDF 파일의 텍스트 데이터를 담은 리스트
    """

    with open(pdf_file_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        text_data = []
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text = page.extract_text()
            text_data.append(text)
        return text_data

def parse_pdf_to_markdown(pdf_file_path):
    """
    PDF 파일을 파싱하여 마크다운 형태의 텍스트 데이터를 추출하는 함수입니다.

    Args:
        pdf_file_path: PDF 파일의 경로

    Returns:
        PDF 파일의 마크다운 형태 텍스트 데이터를 담은 리스트
    """

    with open(pdf_file_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        text_data = []
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text = page.extract_text()
            # 텍스트를 마크다운 형태로 변환
            markdown_text = convert_pdf_to_markdown(text) #text.replace('\n', '\n\n')  # 줄 바꿈을 두 줄로 변환
            text_data.append(markdown_text)
        return text_data

def check_consistency(text_data):
    """
    PDF 파일의 텍스트 데이터의 정합성을 검사하는 함수입니다.

    Args:
        text_data: PDF 파일의 텍스트 데이터를 담은 리스트

    Returns:
        정합성 검사 결과를 담은 딕셔너리
    """

    consistency_results = {}
    error_count = 0

    for page_num, page_text in enumerate(text_data):
        match_results = re.findall(r'제(\d+)조', page_text)
        if match_results:
            if len(match_results) == 1:
                consistency_results[f'페이지 {page_num + 1}'] = {
                    '조항 번호': match_results,
                    '정합성': '정상',
                    '오류 근거': ''
                }
            else:
                consistency_results[f'페이지 {page_num + 1}'] = {
                    '조항 번호': match_results,
                    '정합성': '오류',
                    '오류 근거': '조항 번호가 여러 개 존재합니다.'
                }
                error_count += 1
        else:
            consistency_results[f'페이지 {page_num + 1}'] = {
                '조항 번호': [],
                '정합성': '오류',
                '오류 근거': '조항 번호가 없습니다.'
            }
            error_count += 1

    return consistency_results, error_count

if __name__ == '__main__':
    pdf_file_path = './대한민국헌법(헌법)(제00010호)(19880225).pdf'
    text_data = parse_pdf_to_text(pdf_file_path)
    consistency_results, error_count = check_consistency(text_data)

    print("PDF 파일 텍스트 데이터:")
    for page_num, page_text in enumerate(text_data):
        print(f"페이지 {page_num + 1}: {page_text}")

    print("\n정합성 검사 결과:")
    for page, result in consistency_results.items():
        print(f"{page}: {result}")

    print(f"\n'정합성': '오류' 개수: {error_count}")
    
    text_data = parse_pdf_to_markdown(pdf_file_path)
    print(text_data)
```
